```python
tpl4 = ("Apple", "Banana", "Orange")
print("before:",id(tpl4))
tpl5 = ("kiwi",)
tpl6 = tpl4 + tpl5
tpl4 += tpl5
print(tpl6)
print(tpl4)
print("after:",id(tpl4))
```

```python
tpl4 = ["Apple", "Banana", "Orange"]
print("before:",id(tpl4))
tpl5 = ["kiwi",]
tpl6 = tpl4 + tpl5
tpl4 += tpl5
print(tpl6)
print(tpl4)
print("after:",id(tpl4))
```

```python
tpl7 = ("apple","orange","cherry")
tpl7 = tpl7 + ("mango",)
print(tpl7)
```

```python
tpl8 = ("apple","orange","cherry")
tpl8 = tpl8[:1] + ("mango",) + tpl8[1:]
print(tpl8)
```

```python
tpl9 = ("apple","orange","banana")
tpl10 = ["kiwi","mango"]
tpl9 = tpl9 + tuple(tpl10)
print(tpl9)
print(tpl10)
```
